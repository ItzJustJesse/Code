```python
import requests
from PIL import Image
from io import BytesIO
from transformers import ViTFeatureExtractor, ViTForImageClassification

import matplotlib.pyplot as plt
import numpy as np

# import image modules
import os
```

```python
# Get url with only the best images
UTKFace_url = "best_faces"

# Get total amount of images
image_amount = len(os.listdir(UTKFace_url))
print("Amount of good images:", image_amount)

images = []
# Get images from UTKFace dataset with age range 14 - 17
for image in os.listdir(UTKFace_url):
    # Create list with all images
    images.append(Image.open(UTKFace_url + "/" + image))
```

```python
# Init model, transforms
model = ViTForImageClassification.from_pretrained('nateraw/vit-age-classifier')
transforms = ViTFeatureExtractor.from_pretrained('nateraw/vit-age-classifier')
```

```python
# Transform our image and pass it through the model
inputs = transforms(images, return_tensors='pt')
output = model(**inputs)
```

```python
# Predicted Class probabilities
proba = output.logits.softmax(1)
```

```python
# Predicted Classes
preds = proba.argmax(1)
preds = preds.numpy()
preds
```

```python
age_category, counts = np.unique(preds, return_counts=True)
```

Show the total value count of the classified ages label
- 0 = 0 - 2 years old
- 1 = 3 - 9 years old
- 2 = 10 - 19 years old
- 3 = 20 - 29 years old
- 4 = 30 - 39 years old
- 5 = 40 - 49 years old
- 6 = 50 - 59 years old
- 7 = 60 - 69 years old
- 8 = 70 - 116 years old

```python
plt.bar(age_category, counts)
plt.title("Classified ages of raw faces")
plt.xticks([0,1,2,3,4,5,6,7,8], ["0-2", "3-9", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-116"])
plt.xlabel("Age ranges")
plt.ylabel('Value count of age ranges')
plt.show()
```

```python
# Get url with only the best images
UTKFace_url = "faces_masks"

# Get total amount of images
image_amount = len(os.listdir(UTKFace_url))
print("Amount of good images:", image_amount)

images = []
# Get images from UTKFace dataset with age range 14 - 17
for image in os.listdir(UTKFace_url):
    # Create list with all images
    images.append(Image.open(UTKFace_url + "/" + image))


# Transform our image and pass it through the model
inputs = transforms(images, return_tensors='pt')
output = model(**inputs)


# Predicted Class probabilities
proba = output.logits.softmax(1)


# Predicted Classes
preds = proba.argmax(1)
preds = preds.numpy()
age_category, counts = np.unique(preds, return_counts=True)


plt.bar(age_category, counts)
plt.title("Classified ages of faces with face masks")
plt.xticks([0,1,2,3,4,5,6,7,8], ["0-2", "3-9", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-116"])
plt.xlabel("Age ranges")
plt.ylabel('Value count of age range')
plt.show()
```
