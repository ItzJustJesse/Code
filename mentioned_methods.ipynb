```python
import matplotlib.pyplot as plt
```

```python
# Pie chart with the mentioned adversarial attack method with the total amount it was mentioned
labels = "Sunglasses", "Face masks", "Helmets"
sizes = [4, 3, 1]

# Creates percentages with total amount of mentions
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.1f}% - {v:d}'.format(p=pct,v=val)
    return my_autopct

fig, ax = plt.subplots()
ax.pie(sizes, autopct=make_autopct(sizes), startangle=90, )
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# plt.title("Total amount of mentioned methods")
plt.legend(labels, loc=6)
plt.show()
```
